## Classification with Gemma

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model_id = "google/gemma-2b"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", torch_dtype=torch.float16)
model.eval()
```

```python
def build_prompt(shots=('joy', 'sadness'), prompt_index=1):
    sample_pool = {
        'joy': 'My first child was born.',
        'anger': 'My husband missed an important call, because his phone was on silent AGAIN!',
        'disgust': 'I saw mouldy food.',
        'sadness': 'My dog died last week.',
        'love': 'I told my partner I loved them.',
        'fear': 'I was confronted by a thief.',
        'excitement': 'I got an A in my exam!'
    }

    # Filter and prepare few-shot examples from the selected emotions
    selected_shots = [(sample_pool[e], e) for e in shots if e in sample_pool]

    # Targeted emotions as a string
    emotion_list = "Consider this list of emotions: joy, anger, disgust, sadness, love, fear, excitement. "

    # Prompt templates
    templates = [
        "What are the inferred emotions in the following contexts?",
        emotion_list + "What are the inferred emotions in the following contexts?",
        "",
        "Guess the emotion.",
        "Decipher the emotion from the following statements: ",
        "Decipher the label for the following statements: ",
        "What is the label, for the statement? ",
        "What is the label, given the context? ",
        emotion_list + "Decipher the emotion from the following statements: ",
        emotion_list + "Decipher the label for the following statements: ",
    ]

    header = templates[prompt_index]
    body = header.strip()

    for text, emotion in selected_shots:
        body += f" Context: {text} Answer: {emotion}"

    return lambda x: f"{body} Context: {x} Answer:"

```

```python
from tqdm import tqdm

def classify_with_gemma(ds, tokenizer, model, targeted_emotions, prompt_index=1, shot_emotions=None):
    if shot_emotions is None:
        shot_emotions = targeted_emotions[:7]  # Default to first 3 if not provided

    prompt_func = build_prompt(shots=shot_emotions, prompt_index=prompt_index)
    preds = []

    for example in tqdm(ds, desc="Classifying with Gemma (few-shot)"):
        text = example['text']
        prompt = prompt_func(text)

        inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(model.device)
        with torch.no_grad():
            outputs = model.generate(**inputs, max_new_tokens=3, do_sample=False, pad_token_id=tokenizer.eos_token_id)

        decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
        #print(decoded)
        pred = decoded[len(prompt):].strip().split()[0].lower()

        if pred in targeted_emotions:
            preds.append(targeted_emotions.index(pred))
        else:
            preds.append("other")


    return preds

```

```python
preds_model = classify_with_gemma(ds, tokenizer, model, targeted_emotions, prompt_index=1)
```

```python
from collections import defaultdict
import plotly.graph_objects as go

def plot_sankey(true_indices, pred_indices, targ_em, label_names):
    # Map indices to label names
    true_labels = [label_names[i] for i in true_indices]
    pred_labels = [label_names[i] for i in pred_indices]

    # Unique emotions used in this run
    all_emotions = sorted(set(true_labels + pred_labels))

    # Define node labels
    label_list = all_emotions + [f"pred_{l}" for l in all_emotions]
    label_idx = {l: i for i, l in enumerate(label_list)}

    # Emotion-to-color map
    emotion_to_color = {
        'joy': "#FF0000",
        'anger': "#FFA500",
        'sadness': "#FFFF00",
        'fear': "#27ae60",
        'disgust': "#7f8c8d",
        'love': "#e91e63",
        'excitement': "#f39c12",
        'neutral': "#bdc3c7"
    }

    # Construct node colors: original + pred_ versions
    node_colors = [emotion_to_color.get(e, "#CCCCCC") for e in all_emotions]
    node_colors += node_colors  # repeat for prediction nodes

    # Count transition values
    counter = defaultdict(int)
    for t, p in zip(true_labels, pred_labels):
        src = label_idx[t]
        tgt = label_idx[f"pred_{p}"]
        counter[(src, tgt)] += 1

    # Build link data
    link_source, link_target, link_value, link_color = [], [], [], []

    all_values = list(counter.values())
    max_val = max(all_values)

    for (src, tgt), val in counter.items():
        link_source.append(src)
        link_target.append(tgt)
        link_value.append(val)

        # Normalize alpha to [0.2, 1.0] based on value
        norm_alpha = min(1.0, max(0.2, val / max_val))

        # Convert HEX to RGBA with dynamic transparency
        hex_color = node_colors[src].lstrip('#')
        r, g, b = tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))
        link_color.append(f'rgba({r},{g},{b},{norm_alpha:.2f})')

    # Plot the Sankey
    fig = go.Figure(go.Sankey(
        arrangement="snap",
        node=dict(
            pad=15,
            thickness=25,
            line=dict(color="black", width=0.5),
            label=label_list,
            color=node_colors
        ),
        link=dict(
            source=link_source,
            target=link_target,
            value=link_value,
            color=link_color,
        )
    ))

    fig.update_layout(
        title_text="Emotion Classification Flow (True → Predicted)",
        font_size=13,
        margin=dict(l=40, r=40, t=50, b=40),
        width=600,
        height=600
    )

    fig.show()



```

```python
import gc
import plotly.express as px
from tqdm import tqdm
from sklearn.decomposition import PCA

def visualize_emotion_latent_geometry_from_texts(
    texts, labels, sae, llm_model, tokenizer,
    selected_latents,
    go_emotion_labels=None,
    threshold=1.0,
    max_samples=1000,
    batch_size=8,
    max_length=128,
    device=None
):
    """
    Визуализирует геометрию SAE-латентов на основе текста и известных меток эмоций через PCA.

    Args:
        texts: список текстов
        labels: список меток эмоций (индексы или строки), один к одному с texts
        sae: объект SAE с .W_dec и конфигурацией
        llm_model: модель (HookedSAETransformer) с методом run_with_cache_with_saes
        tokenizer: соответствующий токенизатор
        selected_latents: список индексов SAE-латентов
        go_emotion_labels: словарь отображения {int: str} для меток
        threshold: порог активации выбранных латентов
        max_samples: макс. число точек для визуализации
        batch_size: размер батча для ускорения
        max_length: макс. длина токенов для обрезки
        device: torch.device (если None — определяется автоматически)
    """
    if device is None:
        device = next(llm_model.parameters()).device

    reconstructions = []
    final_labels = []
    tokens = []
    contexts = []

    for i in tqdm(range(0, len(texts), batch_size), desc="Processing batches"):
        gc.collect()
        torch.cuda.empty_cache()

        batch_texts = texts[i:i + batch_size]
        batch_labels = labels[i:i + batch_size]

        tokenized = tokenizer(
            batch_texts,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=max_length
        ).to(device)

        input_ids = tokenized["input_ids"]

        # Forward pass and capture SAE activations
        _, cache = llm_model.run_with_cache_with_saes(
            input_ids,
            saes=[sae],
            stop_at_layer=sae.cfg.hook_layer + 1,
            names_filter=[f"{sae.cfg.hook_name}.hook_sae_acts_post"],
        )

        # Extract full SAE activations: shape [B, T, F]
        sae_acts = cache[f"{sae.cfg.hook_name}.hook_sae_acts_post"].detach().cpu()

        # Loop over batch samples
        for b in range(sae_acts.shape[0]):
            label = batch_labels[b]
            text = batch_texts[b]

            # Option: select last token's SAE activation
            token_act = sae_acts[b, -1]  # shape: [F]

            if (token_act[selected_latents] > threshold).any():
                selected_acts = token_act[selected_latents]
                recon = sae.W_dec[selected_latents].T.to(selected_acts.device) @ selected_acts

                reconstructions.append(recon.numpy())
                tokens.append(text[:40])
                contexts.append(text)
                readable_label = (
                    go_emotion_labels.get(label, str(label)) if go_emotion_labels else str(label)
                )
                final_labels.append(readable_label)

            if len(reconstructions) >= max_samples:
                break
        if len(reconstructions) >= max_samples:
            break

    if not reconstructions:
        print("⚠️ No active reconstructions found.")
        return

    # PCA on reconstructions
    X = torch.tensor(reconstructions).numpy()
    pca = PCA(n_components=3)
    X_pca = pca.fit_transform(X)

    # DataFrame for plotting
    df = pd.DataFrame({
        "PC1": X_pca[:, 0],
        "PC2": X_pca[:, 1],
        #"PC3": X_pca[:, 2],
        "emotion": final_labels,
        "token": tokens,
        "context": contexts,
    })

    fig = px.scatter(
        df,
        #x="PC1",
        x="PC1",
        y="PC2",
        color="emotion",
        hover_data=["context"],
        hover_name="token",
        height=700,
        width=1000,
        title="PCA Geometry of Emotion-Selective SAE Reconstructions (Gemma)",
        color_discrete_sequence=px.colors.qualitative.Set3
    )
    fig.show()


```

```python
import torch
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from tqdm import tqdm
import gc

def visualize_emotion_latent_geometry(
    texts,
    labels,
    sae,
    llm_model,
    tokenizer,
    selected_latents,
    go_emotion_labels=None,
    threshold=1.0,
    max_samples=1000,
    batch_size=8,
    max_length=128,
    mode="pca",  # "pca" or "tsne"
    device=None,
    tsne_perplexity=30,
    tsne_learning_rate=200
):
    """
    Визуализирует геометрию SAE латентов (PCA или t-SNE) на основе текста и эмоций.

    Args:
        mode: "pca" (default) or "tsne"
    """
    if device is None:
        device = next(llm_model.parameters()).device

    all_data = {"recons": [], "context": [], "token": [], "token_group": []}
    collected = 0

    for i in tqdm(range(0, len(texts), batch_size), desc="Processing batches"):
        if collected >= max_samples:
            break

        gc.collect()
        torch.cuda.empty_cache()

        batch_texts = texts[i:i + batch_size]
        batch_labels = labels[i:i + batch_size]

        tokenized = tokenizer(
            batch_texts,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=max_length
        ).to(device)

        input_ids = tokenized["input_ids"]
        attention_mask = tokenized.get("attention_mask", None)

        _, cache = llm_model.run_with_cache_with_saes(
            input_ids,
            saes=[sae],
            stop_at_layer=sae.cfg.hook_layer + 1,
            names_filter=[f"{sae.cfg.hook_name}.hook_sae_acts_post"],
        )

        sae_acts = cache[f"{sae.cfg.hook_name}.hook_sae_acts_post"][..., selected_latents].detach().cpu()
        full_seq_len = sae_acts.shape[1]
        batch_input_ids = input_ids.detach().cpu()

        for b in range(sae_acts.shape[0]):
            for t in range(sae_acts.shape[1]):
                act = sae_acts[b, t]

                if (act > threshold).any():
                    recon = sae.W_dec[selected_latents].T.cpu() @ act
                    all_data["recons"].append(recon.numpy())

                    token = tokenizer.decode(batch_input_ids[b, t])
                    label = batch_labels[b]
                    readable = go_emotion_labels.get(label, str(label)) if go_emotion_labels else str(label)

                    left = max(t - 5, 0)
                    right = min(t + 6, full_seq_len)
                    context = tokenizer.decode(batch_input_ids[b, left:right])

                    all_data["token"].append(token)
                    all_data["context"].append(context)
                    all_data["token_group"].append(readable)

                    collected += 1
                    if collected >= max_samples:
                        break
            if collected >= max_samples:
                break

    if not all_data["recons"]:
        print("⚠️ No reconstructions found.")
        return

    X = torch.tensor(all_data.pop("recons")).numpy()

    # Choose PCA or t-SNE
    if mode == "pca":
        reducer = PCA(n_components=3)
        embedding = reducer.fit_transform(X)
        all_data["PC2"] = embedding[:, 1]
        all_data["PC3"] = embedding[:, 2]
        title = "PCA Geometry of Emotion-Selective SAE Reconstructions (Gemma)"
    elif mode == "tsne":
        reducer = TSNE(n_components=2, perplexity=tsne_perplexity, learning_rate=tsne_learning_rate, init="pca")
        embedding = reducer.fit_transform(X)
        all_data["PC2"] = embedding[:, 0]
        all_data["PC3"] = embedding[:, 1]
        title = "t-SNE Projection of Emotion-Selective SAE Reconstructions (Gemma)"
    else:
        raise ValueError("mode must be 'pca' or 'tsne'")

    df = pd.DataFrame(all_data)

    fig = px.scatter(
        df,
        x="PC2",
        y="PC3",
        color="token_group",
        hover_data=["context"],
        hover_name="token",
        height=700,
        width=1000,
        title=title,
        color_discrete_sequence=px.colors.qualitative.Set3,
    )
    fig.show()

```

```python
from tqdm import tqdm
import torch
import gc

batch_size = 4
max_length = 40
activation_threshold = 1.0
targ_neurons = set(targ_neurons)
results = []

for i in tqdm(range(0, len(texts), batch_size), desc="Processing batches"):

        gc.collect()
        torch.cuda.empty_cache()

        batch_texts = texts[i:i + batch_size]

        tokenized = tokenizer(
            batch_texts,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=max_length
        )
        input_ids = tokenized["input_ids"].to(device)

        _, cache = gemma.run_with_cache_with_saes(
            input_ids,
            saes=[gemma_sae],
            stop_at_layer=gemma_sae.cfg.hook_layer + 1,
            names_filter=[f"{gemma_sae.cfg.hook_name}.hook_sae_acts_post"],
        )

        sae_acts = cache[f"{gemma_sae.cfg.hook_name}.hook_sae_acts_post"].detach().cpu()  # [B, T, F]
        final_acts = sae_acts[:, -1, :]  # [B, F]
        decoded_tokens = [tokenizer.convert_ids_to_tokens(ids) for ids in input_ids]

        for j in range(len(batch_texts)):
            token_acts = sae_acts[j]  # [T, F]
            filtered_entries = []

            for t in range(token_acts.shape[0]):  # iterate over token positions
                for f in targ_neurons:
                    act_val = token_acts[t, f].item()
                    if act_val > activation_threshold:
                        filtered_entries.append((t, f, act_val))

            final_filtered = {
                f: final_acts[j, f].item()
                for f in targ_neurons
                if final_acts[j, f].item() > activation_threshold
            }

            results.append({
                "input_ids": input_ids[j].detach().cpu(),
                "tokens": decoded_tokens[j],
                "activation_targeted": final_filtered,
                "active_neurons": filtered_entries,
                "sparsity": len(filtered_entries),
            })

        del cache, sae_acts, final_acts, input_ids, tokenized
        torch.cuda.empty_cache()
```

```python
from tqdm import tqdm
import torch
import gc

batch_size = 4
max_length = 400
results = []

for i in tqdm(range(0, len(texts), batch_size), desc="Processing batches"):
    try:
        gc.collect()
        torch.cuda.empty_cache()

        # Get batch
        batch_texts = texts[i:i + batch_size]

        # Tokenize
        tokenized = tokenizer(
            batch_texts,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=max_length
        )
        input_ids = tokenized["input_ids"].to(device)

        # Forward pass with SAE and cache activations
        _, cache = gemma.run_with_cache_with_saes(
            input_ids,
            saes=[gemma_sae],
            stop_at_layer=gemma_sae.cfg.hook_layer + 1,
            names_filter=[f"{gemma_sae.cfg.hook_name}.hook_sae_acts_post"],
        )

        # SAE activations (features)
        sae_acts = cache[f"{gemma_sae.cfg.hook_name}.hook_sae_acts_post"]  # [B, T, F]
        final_acts = sae_acts[:, -1, :].detach().cpu()

        # Sparsity = number of active features
        sparsity = (sae_acts[:, -1, :] > 1).sum(dim=-1)  # [B]

        decoded_tokens = [tokenizer.convert_ids_to_tokens(ids) for ids in input_ids]


        for i in range(len(batch_texts)):
            results.append({
                "input_ids": input_ids[i].detach().cpu(),              # torch.Tensor
                "tokens": decoded_tokens[i],                          # list of strings
                "activation": final_acts[i],                          # torch.Tensor
                "sparsity": int(sparsity[i]),                         # int
            })

        # Cleanup
        del cache, sae_acts, final_acts, sparsity, input_ids, tokenized
        torch.cuda.empty_cache()

    except RuntimeError as e:
        print(f"⚠️ OOM on batch {i}-{i+batch_size}: {e}")
        torch.cuda.empty_cache()
```

```python
from tqdm import tqdm
import torch
import gc

batch_size = 4
max_length = 40
activation_threshold = 1.0
targ_neurons = set(targ_neurons)
results = []

for i in tqdm(range(0, len(texts), batch_size), desc="Processing batches"):

        gc.collect()
        torch.cuda.empty_cache()

        batch_texts = texts[i:i + batch_size]

        tokenized = tokenizer(
            batch_texts,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=max_length
        )
        input_ids = tokenized["input_ids"].to(device)

        _, cache = gemma.run_with_cache_with_saes(
            input_ids,
            saes=[gemma_sae],
            stop_at_layer=gemma_sae.cfg.hook_layer + 1,
            names_filter=[f"{gemma_sae.cfg.hook_name}.hook_sae_acts_post"],
        )

        sae_acts = cache[f"{gemma_sae.cfg.hook_name}.hook_sae_acts_post"].detach().cpu()  # [B, T, F]
        final_acts = sae_acts[:, -1, :]  # [B, F]
        decoded_tokens = [tokenizer.convert_ids_to_tokens(ids) for ids in input_ids]

        for j in range(len(batch_texts)):
            token_acts = sae_acts[j]  # [T, F]
            filtered_entries = []

            for t in range(token_acts.shape[0]):  # iterate over token positions
                for f in targ_neurons:
                    act_val = token_acts[t, f].item()
                    if act_val > activation_threshold:
                        filtered_entries.append((t, f, act_val))

            final_filtered = {
                f: final_acts[j, f].item()
                for f in targ_neurons
                if final_acts[j, f].item() > activation_threshold
            }

            results.append({
                "input_ids": input_ids[j].detach().cpu(),
                "tokens": decoded_tokens[j],
                "activation_targeted": final_filtered,
                "active_neurons": filtered_entries,
                "sparsity": len(filtered_entries),
            })

        del cache, sae_acts, final_acts, input_ids, tokenized
        torch.cuda.empty_cache()
```

# Steering

```python
import torch
from sae_lens import SAE, HookedSAETransformer

def steer_emotion_gemma_sae(
    hidden: torch.Tensor,
    sae: SAE,
    emotion_feature_ids: list[int],
    method: str = "amplify",
    alpha: float = 1.0,
    beta: float = 1.0,
) -> torch.Tensor:
    """
    Steer the hidden representation by enhancing emotion-related SAE features.

    Args:
        hidden: Tensor of shape (seq_len, hidden_dim) from a selected layer.
        sae: Pretrained SAE from sae_lens (with encode() and W).
        emotion_feature_ids: List of indices of emotion-related SAE features.
        method: "amplify" or "calibrate".
        alpha: Scaling factor for amplification.
        beta: Scaling factor for calibration.

    Returns:
        Steered hidden representation (same shape as input).
    """
    # 1. Encode into SAE activation space
    A = sae.encode(hidden)  # shape: (seq_len, num_features)

    # 2. Select emotion-related features
    S = sae.W[emotion_feature_ids]              # shape: (num_emotion_features, hidden_dim)
    A_emotion = A[:, emotion_feature_ids]       # shape: (seq_len, num_emotion_features)

    # 3. Modify the hidden representation
    if method == "amplify":
        delta = alpha * (A_emotion @ S)         # shape: (seq_len, hidden_dim)
        return hidden + delta

    elif method == "calibrate":
        mean_feature = S.mean(dim=0)            # shape: (hidden_dim,)
        projection = A_emotion @ S              # shape: (seq_len, hidden_dim)
        return hidden - projection + beta * mean_feature

    else:
        raise ValueError("Invalid method: choose 'amplify' or 'calibrate'")
```

```python
# Define the patching function
def patch_activations_amplify(cache):
    # Get the SAE activations [batch, seq_len, latent_dim]
    sae_acts = cache[hook_name]  # tensor

    # Clone to avoid in-place ops
    patched_acts = sae_acts.clone()

    # ReLU portion of only anger-related features
    patched_acts[..., anger_neurons] += alpha * torch.relu(sae_acts[..., anger_neurons])

    return {hook_name: patched_acts}

# Input prompt
prompt = "I can't believe what just happened."

# Tokenize
inputs = tokenizer(prompt, return_tensors="pt").to(device)

# Run inference with activation patching
patched_logits = gemma.run_with_hooks(
    inputs["input_ids"],
    fwd_hooks=[
        (hook_name, lambda v, hook: patch_activations_amplify({hook_name: v})[hook_name])
    ],
    return_type="logits"
)

# Decode result
patched_output = tokenizer.decode(patched_logits.argmax(dim=-1)[0], skip_special_tokens=True)
print("🔁 Output after anger amplification:\n", patched_output)
```

```python
from tqdm import tqdm
from functools import partial
import torch

# === Define your anger neuron index ===
anger_neuron_idx = 2438  # replace with actual anger-selective neuron index

# === Function to find max activation of this neuron ===
def find_max_activation(model, sae, act_store, neuron_idx, num_batches=100):
    max_activation = 0.0
    pbar = tqdm(range(num_batches), desc="Finding max activation")

    for _ in pbar:
        tokens = act_store.get_batch_tokens()
        _, cache = model.run_with_cache_with_saes(
            tokens,
            saes=[sae],
            stop_at_layer=sae.cfg.hook_layer + 1,
            names_filter=[f"{sae.cfg.hook_name}.hook_sae_acts_post"]
        )
        acts = cache[f"{sae.cfg.hook_name}.hook_sae_acts_post"]
        acts_flat = acts.flatten(0, 1)
        batch_max = acts_flat[:, neuron_idx].max().item()
        max_activation = max(max_activation, batch_max)
        pbar.set_description(f"Max activation: {max_activation:.2f}")

    return max_activation

# === Hook for steering ===
def steering_hook_fn(resid_pre, hook, steering_vector, strength, max_act):
    return resid_pre + max_act * strength * steering_vector

# === Generate with steering ===
def generate_with_steering(model, sae, prompt, neuron_idx, max_act, strength=1.0, max_new_tokens=10):
    input_ids = model.to_tokens(prompt, prepend_bos=sae.cfg.prepend_bos)

    steer_vec = sae.W_dec[neuron_idx].to(model.cfg.device)

    hook_fn = partial(
        steering_hook_fn,
        steering_vector=steer_vec,
        strength=strength,
        max_act=max_act
    )

    with model.hooks(fwd_hooks=[(sae.cfg.hook_name, hook_fn)]):
        output_ids = model.generate(
            input_ids,
            max_new_tokens=max_new_tokens,
            temperature=0.7,
            top_p=0.9,
            stop_at_eos=True,
            prepend_bos=sae.cfg.prepend_bos
        )

    return model.tokenizer.decode(output_ids[0])

# === Run the workflow ===

# Step 1: Get max activation of anger neuron
max_act = find_max_activation(gemma, gemma_sae, activation_store, anger_neuron_idx)
print(f"\n🔧 Max activation for anger neuron {anger_neuron_idx}: {max_act:.2f}")

# Step 2: Prompt
prompt = "I can't believe you said that to me"

# Step 3: Generate baseline
print("\n📝 Without steering:")
baseline = gemma.generate(prompt, max_new_tokens=10, stop_at_eos=True)
print(baseline)

# Step 4: Generate with anger steering
print("\n🔥 With anger neuron steering:")
steered = generate_with_steering(gemma, gemma_sae, prompt, anger_neuron_idx, max_act, strength=2.0)
print(steered)
```
